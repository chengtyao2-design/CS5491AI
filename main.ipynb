```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取程序生成的 CSV 文件
# 假设你的文件叫 score_log.csv，第一列是迭代次数(Iteration)，第二列是最高分(Score)
try:
    df = pd.read_csv('score_log.csv', names=['Iteration', 'Score'])
    
    # 2. 帮我保存到 JSON 文件中
    # orient='records' 会把它存成列表字典的格式，非常清晰
    df.to_json('score_log.json', orient='records', indent=4)
    print("✅ 数据已成功保存到当前目录下的 score_log.json！\n")

    # 3. 绘制折线图
    plt.figure(figsize=(10, 6)) # 设置画布大小
    
    # 画折线，marker='o' 表示在每个数据点画个小圆圈
    plt.plot(df['Iteration'], df['Score'], marker='o', linestyle='-', color='b', label='Global Best Score')
    
    # 设置图表标题和坐标轴标签
    plt.title('FunSearch Optimization Progress', fontsize=14)
    plt.xlabel('Iteration', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    
    # 开启网格线，方便查看分数对应的位置
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend() # 显示图例
    
    # 显示图表
    plt.show()

except FileNotFoundError:
    print("❌ 找不到 score_log.csv 文件，请确认你的程序已经跑出数据并生成了该文件！")
```
